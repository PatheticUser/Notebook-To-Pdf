# Imports


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This command allows plots to appear directly in the notebook
%matplotlib inline
```

#  Sample Data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [150, 200, 180, 220, 250, 210],
    'Expenses': [80, 90, 85, 110, 120, 105]
}

df = pd.DataFrame(data)
df['Profit'] = df['Sales'] - df['Expenses']

# Display the first few rows
df.head()
```

#  Basic Data Analysis

```python
print(f"Total Profit for the half-year: ${df['Profit'].sum()}")
print(f"Average Monthly Sales: ${df['Sales'].mean():.2f}")
```
